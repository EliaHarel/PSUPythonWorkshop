```python
import matplotlib.pyplot as plt
import matplotlib.cm as brewer
import numpy as np
year, universityPark, otherLocs, worldCampus, totPSU = np.loadtxt('pennStateEnroll.dat', unpack=True, comments='#', skiprows=2)

set2 = brewer.get_cmap('autumn',3)
figPSU, ax = plt.subplots()
ax.plot( year, universityPark, label = 'University Park', color=set2(0))
ax.plot( year, otherLocs, label = 'Branch Campuses', color=set2(1))
ax.plot( year, worldCampus, label = 'World Campus', color=set2(2))
###################################################
ax.set_xlabel( 'Year' , fontsize=18)
ax.set_ylabel( 'Enrollment' , fontsize=18)
###################################################
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.text(2020, universityPark[0], 'University Park', horizontalalignment='left', verticalalignment='center')
ax.text(2020, otherLocs[0], 'Branch Campuses', horizontalalignment='left', verticalalignment='center')
ax.text(2020, worldCampus[0], 'World Campuses', horizontalalignment='left', verticalalignment='center')

#ax.legend( loc = 2 )
plt.show(figPSU)
```

```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True) # initiate notebook for offline plot

import plotly.graph_objs as go
import numpy as np
year, universityPark, otherLocs, worldCampus, totPSU = np.loadtxt('pennStateEnroll.dat', unpack=True, comments='#', skiprows=2)

# Create traces
trace0 = go.Scatter(
    x = year,
    y = universityPark,
    mode = 'lines',
    name = 'University Park'
)
trace1 = go.Scatter(
    x = year,
    y = otherLocs,
    mode = 'lines+markers',
    name = 'Branch Campuses'
)
trace2 = go.Scatter(
    x = year,
    y = worldCampus,
    mode = 'markers',
    name = 'World Campuses'
)
data = [trace0, trace1, trace2]

iplot(data)
```

```python
import altair as alt
import pandas as pd

year, universityPark, otherLocs, worldCampus, totPSU = np.loadtxt('pennStateEnroll.dat', unpack=True, comments='#', skiprows=2)
d = {'year': year, 'universityPark': universityPark, 'otherLocs': otherLocs, 'worldCampus': worldCampus}
df = pd.DataFrame(data=d)  
print df.head()
```

```python
%%javascript
element.append('<div id="graph" style="min-width: 800px; height: 400px; margin: 0 auto"></div>');
require.config({paths: {d3: "//d3js.org/d3.v4.min",}});

require(["d3"], function(d3) {
    console.log(data)
    // Use this to keep the screen clean
    d3.select("div#graph").selectAll("*").remove();
    
    // declare width/height
    var width = 800, height = 400, margin = 50;
    //define the x range/domain
    var x = d3.scaleLinear().range([50, width-50]).domain(d3.extent(data[0], d=>d[0]));
    //we need to find the max/min for each of loaation
    const extentUP = d3.extent(data[0], d=>d[1])
    const extentOL = d3.extent(data[1], d=>d[1])
    const extentWC = d3.extent(data[2], d=>d[1])
    
    //define the y range/domain
    var y = d3.scaleLinear().range([height-margin, margin])
    y.domain(d3.extent([extentUP[0],extentUP[1],extentOL[0],extentOL[1],extentWC[0],extentWC[1]]));
    
    // 7. d3's line generator
    var line = d3.line()
        .x(function(d) { return x(d[0]); }) // set the x values for the line generator
        .y(function(d) { return y(d[1]); }) // set the y values for the line generator 
        .curve(d3.curveMonotoneX) // apply smoothing to the line
    var color = d3.scaleOrdinal(d3.schemeCategory10); // Additional details on Color: https://github.com/d3/d3-scale-chromatic
    // Same as example 5, just condensed 
    var svg = d3.select("div#graph").append("svg")
        .attr("width", width+"px")
        .attr("height", height+"px");
    
    // 3. Call the x axis in a group tag
    svg.append("g")
        .attr("class", "x axis")
        .attr("transform", "translate(0," + (height-margin) + ")")
        .call(d3.axisBottom(x)); // Create an axis component with d3.axisBottom

    // 4. Call the y axis in a group tag
    svg.append("g")
        .attr("class", "y axis")
        .attr("transform", "translate("+ (margin) + ",0)")
        .call(d3.axisLeft(y)); // Create an axis component with d3.axisLeft

    var graphPlacement = svg.selectAll("g.plots").data(data)
        .enter().append("g")
        .attr("class", "plots")

    graphPlacement.each(function(d,i){
        svg.append("path")
            .datum(d) // 10. Binds data to the line 
            .attr("class", "line") // Assign a class for styling 
            .attr("d", line) // 11. Calls the line generator 
            .style("fill","none")
            .style("stroke",color(i))      

        svg.selectAll("circle.data"+i).data(d)
            .enter().append("circle") // Uses the enter().append() method
            .attr("class", "data"+i) // Assign a class for styling
            .attr("cx", function(d) {return x(d[0]) })
            .attr("cy", function(d) { return y(d[1]) })
            .attr("r", 2)
            .style("stroke",color(i))
            .style("fill",color(i))
                        
    })
    

    
})

```
