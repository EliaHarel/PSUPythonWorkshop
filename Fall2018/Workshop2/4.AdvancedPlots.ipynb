## Other methods to visualize data

### Plotly

plotly uses a pretty straight-forward approach to graphing. Map values to the x,y coordinate. 

```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True) # initiate notebook for offline plot

import plotly.graph_objs as go
import numpy as np
year, universityPark, otherLocs, worldCampus, totPSU = np.loadtxt('pennStateEnroll.dat', unpack=True, comments='#', skiprows=2)

# Create traces
trace0 = go.Scatter(
    x = year,
    y = universityPark,
    mode = 'lines',
    name = 'University Park'
)
trace1 = go.Scatter(
    x = year,
    y = otherLocs,
    mode = 'lines+markers',
    name = 'Branch Campuses'
)
trace2 = go.Scatter(
    x = year,
    y = worldCampus,
    mode = 'markers',
    name = 'World Campuses'
)
data = [trace0, trace1, trace2]

iplot(data)
```

### Altair

Altair expects a pandas data frame (similar to R)
We will talk about this more next workshop. 

```python
import altair as alt
import pandas as pd

year, universityPark, otherLocs, worldCampus, totPSU = np.loadtxt('pennStateEnroll.dat', unpack=True, comments='#', skiprows=2)
# first, format data to fit a data frame
d = {'year': year, 'universityPark': universityPark, 'otherLocs': otherLocs, 'worldCampus': worldCampus}
# second, create the data frame
df = pd.DataFrame(data=d)  
# let's preview the data
print df.head()
```

Again, similar to R. We will "melt" the data. Here is a quick example used from: https://deparkes.co.uk/2016/10/28/reshape-pandas-data-with-melt/

```python
testData = {'weekday': ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
        'Person 1': [12, 6, 5, 8, 11, 6, 4],
        'Person 2': [10, 6, 11, 5, 8, 9, 12],
        'Person 3': [8, 5, 7, 3, 7, 11, 15]}
print testData, "\n"

testDataFrame = pd.DataFrame(testData, columns=['weekday','Person 1', 'Person 2', 'Person 3'])
print testDataFrame.head(), "\n"

testmelted = pd.melt(testDataFrame, id_vars=["weekday"], var_name="Person", value_name="Score")
print testmelted.head(), "\n"
```

Back to our data.

```python
melted = pd.melt(df, id_vars=["year"], var_name="location", value_name="enrollment")
print melted.head()
```

First, we need to change up the data a bit to fit this D3.js graph. 

This approach uses D3.js (https://d3js.org/). Again, D3.js has a pretty big learning curve, but will enivably provide the most functionality in the long run (also, free/open-source). 

```python
from IPython.display import HTML
from IPython.display import Javascript

year, universityPark, otherLocs, worldCampus, totPSU = np.loadtxt('pennStateEnroll.dat', unpack=True, comments='#', skiprows=2)

# we are creating three bins for the data
allData = []

for i in range(len(year)):
    allData.append([year[i], universityPark[i],"University Park"])
    allData.append([year[i], otherLocs[i],"Other Locations"])
    allData.append([year[i], worldCampus[i],"World Campus"])


for i in allData[0:5]:
    print i
Javascript("""window.data={};""".format(allData))
```

Next we need to create the actual lines/marks (paths/circles)

```python
%%javascript
element.append('<div id="graph" style="min-width: 800px; height: 400px; margin: 0 auto"></div>');
require.config({paths: {d3: "//d3js.org/d3.v4.min",}});

require(["d3"], function(d3) {
    // sorting the data by date
    data.sort(function(a, b){return a[0] - b[0]});
    
    // Use this to keep the screen clean
    d3.select("div#graph").selectAll("*").remove();
    
    // declare width/height
    var width = 800, height = 400, margin = 60;
    
    //define the x range/domain
    var x = d3.scaleLinear().range([50, width-50]).domain(d3.extent(data, d=>d[0]));
    
    //define the y range/domain
    var y = d3.scaleLinear().range([height-margin, margin]).domain(d3.extent(data,d=>d[1]));
    
    // d3's line generator
    var line = d3.line()
        .x(function(d) { return x(d[0]); }) // set the x values for the line generator
        .y(function(d) { return y(d[1]); }) // set the y values for the line generator 
        .curve(d3.curveMonotoneX) // apply smoothing to the line
    var color = d3.scaleOrdinal(d3.schemeCategory10); // Additional details on Color: https://github.com/d3/d3-scale-chromatic
    
    // first, add a container for the SVG object
    var svg = d3.select("div#graph").append("svg")
        .attr("width", width+"px")
        .attr("height", height+"px");
    
    // create the x axis in a group tag
    svg.append("g")
        .attr("class", "x axis")
        .attr("transform", "translate(0," + (height-margin) + ")")
        .call(d3.axisBottom(x)); // Create an axis component with d3.axisBottom
    
    // text for label
    svg.append("text")             
        .attr("transform","translate(" + (width/2) + " ," + (height - 20) + ")")
        .style("text-anchor", "middle")
        .text("Date");
    
    // create the y axis in a group tag
    svg.append("g")
        .attr("class", "y axis")
        .attr("transform", "translate("+ (margin) + ",0)")
        .call(d3.axisLeft(y)) // Create an axis component with d3.axisLeft
    
    // text for label
    svg.append("text")
        .attr("transform", "rotate(-90)")
        .attr("y", 10)
        .attr("x", - (height / 2))
        .style("text-anchor", "middle")
        .text("Value");
    
    // we have 3 plots, so we will group these together using nest
    svg.selectAll("path.line").data(d3.nest().key(function(d) { return d[2]; }).entries(data)) // binds data to the line 
        .enter().append("path")
        .attr("class", "line") // Assign a class for styling 
        .attr("d",d=> line(d.values)) // Calls the line generator 
        .style("fill","none")
        .style("stroke",(d,i)=>color(d.key))  
    
     // adding circle for each mark
    svg.selectAll("circle.point").data(data) // binds data to the line 
        .enter().append("circle")
        .attr("class", "point") // Assign a class for styling 
        .attr("cx", function(d) {return x(d[0]) })
        .attr("cy", function(d) { return y(d[1]) })
        .attr("r", 2)
        .style("stroke",d=>color(d[2]))
        .style("fill",d=>color(d[2]))
})
```

Where D3.js starts to shine is when we utilize it with transitions (animations). 

```python
%%javascript
element.append('<div id="graph2" style="min-width: 800px; height: 400px; margin: 0 auto"></div>');
require.config({paths: {d3: "//d3js.org/d3.v4.min",}});

require(["d3"], function(d3) {
    // sorting the data by date
    data.sort(function(a, b){return a[0] - b[0]});
    
    // Use this to keep the screen clean
    d3.select("div#graph2").selectAll("*").remove();
    
    // declare width/height
    var width = 800, height = 400, margin = 60;
    
    //define the x range/domain
    var x = d3.scaleLinear().range([50, width-50]).domain(d3.extent(data, d=>d[0]));
    
    //define the y range/domain
    var y = d3.scaleLinear().range([height-margin, margin]).domain(d3.extent(data,d=>d[1]));
    
    // d3's line generator
    var line = d3.line()
        .x(function(d) { return x(d[0]); }) // set the x values for the line generator
        .y(function(d) { return y(d[1]); }) // set the y values for the line generator 
        .curve(d3.curveMonotoneX) // apply smoothing to the line
    var color = d3.scaleOrdinal(d3.schemeCategory10); // Additional details on Color: https://github.com/d3/d3-scale-chromatic
    
    // first, add a container for the SVG object
    var svg = d3.select("div#graph2").append("svg")
        .attr("width", width+"px")
        .attr("height", height+"px");
    
    // create the x axis in a group tag
    svg.append("g")
        .attr("class", "x axis")
        .attr("transform", "translate(0," + (height-margin) + ")")
        .call(d3.axisBottom(x)); // Create an axis component with d3.axisBottom
    
    // add text for label
    svg.append("text")             
        .attr("transform","translate(" + (width/2) + " ," + (height - 20) + ")")
        .style("text-anchor", "middle")
        .text("Date");
    
    // create the y axis in a group tag
    svg.append("g")
        .attr("class", "y axis")
        .attr("transform", "translate("+ (margin) + ",0)")
        .call(d3.axisLeft(y)) // Create an axis component with d3.axisLeft
    
    // add text for label
    svg.append("text")
        .attr("transform", "rotate(-90)")
        .attr("y", 10)
        .attr("x", - (height / 2))
        .style("text-anchor", "middle")
        .text("Value");

    // we have 3 plots, so we will group these together
    svg.selectAll("path.line").data(d3.nest().key(function(d) { return d[2]; }).entries(data)) // 10. Binds data to the line 
        .enter().append("path")
        .attr("class", "line") // Assign a class for styling 
        .attr("d",d=> line(d.values)) // Calls the line generator 
        //.attr("d",d=> console.log(d.values)) // 11. Calls the line generator
        .style("fill","none")
        .style("stroke",(d,i)=>color(d.key)) 
        .transition()
        .duration(2000)
        .attrTween("stroke-dasharray", function() {
            var len = this.getTotalLength();
            return function(t) { return (d3.interpolateString("0," + len, len + ",0"))(t) };
           })
    
    // adding circle for each mark
    svg.selectAll("circle.point").data(data) // 10. Binds data to the line 
        .enter().append("circle")
        .attr("class", "point") // Assign a class for styling 
        .attr("cx", function(d) {return x(d[0]) })
        .attr("cy", function(d) { return y(d[1]) })
        .attr("r", 0)
        .style("stroke",d=>color(d[2]))
        .style("fill",d=>color(d[2]))
        .transition()
        .duration(2000)
        .delay((d,i)=>(d[0]-d3.min(data,d=>d[0]))*40)
        .attr("r", 2)
})
```
